````python
examples = [
    {
        "question": "ì§€êµ¬ì˜ ëŒ€ê¸° ì¤‘ ê°€ì¥ ë§ì€ ë¹„ìœ¨ì„ ì°¨ì§€í•˜ëŠ” ê¸°ì²´ëŠ” ë¬´ì—‡ì¸ê°€ìš”?",
        "answer": "ì§€êµ¬ ëŒ€ê¸°ì˜ ì•½ 78%ë¥¼ ì°¨ì§€í•˜ëŠ” ì§ˆì†Œì…ë‹ˆë‹¤."
    },
    {
        "question": "ê´‘í•©ì„±ì— í•„ìš”í•œ ì£¼ìš” ìš”ì†Œë“¤ì€ ë¬´ì—‡ì¸ê°€ìš”?",
        "answer": "ê´‘í•©ì„±ì— í•„ìš”í•œ ì£¼ìš” ìš”ì†ŒëŠ” ë¹›, ì´ì‚°í™”íƒ„ì†Œ, ë¬¼ì…ë‹ˆë‹¤."
    },
    {
        "question": "í”¼íƒ€ê³ ë¼ìŠ¤ ì •ë¦¬ë¥¼ ì„¤ëª…í•´ì£¼ì„¸ìš”.",
        "answer": "í”¼íƒ€ê³ ë¼ìŠ¤ ì •ë¦¬ëŠ” ì§ê°ì‚¼ê°í˜•ì—ì„œ ë¹—ë³€ì˜ ì œê³±ì´ ë‹¤ë¥¸ ë‘ ë³€ì˜ ì œê³±ì˜ í•©ê³¼ ê°™ë‹¤ëŠ” ê²ƒì…ë‹ˆë‹¤."
    },
    {
        "question": "ì§€êµ¬ì˜ ìì „ ì£¼ê¸°ëŠ” ì–¼ë§ˆì¸ê°€ìš”?",
        "answer": "ì§€êµ¬ì˜ ìì „ ì£¼ê¸°ëŠ” ì•½ 24ì‹œê°„(ì •í™•íˆëŠ” 23ì‹œê°„ 56ë¶„ 4ì´ˆ)ì…ë‹ˆë‹¤."
    },
    {
        "question": "DNAì˜ ê¸°ë³¸ êµ¬ì¡°ë¥¼ ê°„ë‹¨íˆ ì„¤ëª…í•´ì£¼ì„¸ìš”.",
        "answer": "DNAëŠ” ë‘ ê°œì˜ í´ë¦¬ë‰´í´ë ˆì˜¤í‹°ë“œ ì‚¬ìŠ¬ì´ ì´ì¤‘ ë‚˜ì„  êµ¬ì¡°ë¥¼ ì´ë£¨ê³  ìˆìŠµë‹ˆë‹¤."
    },
    {
        "question": "ì›ì£¼ìœ¨(Ï€)ì˜ ì •ì˜ëŠ” ë¬´ì—‡ì¸ê°€ìš”?",
        "answer": "ì›ì£¼ìœ¨(Ï€)ì€ ì›ì˜ ì§€ë¦„ì— ëŒ€í•œ ì›ì˜ ë‘˜ë ˆì˜ ë¹„ìœ¨ì…ë‹ˆë‹¤."
    }
]
````

##### FewShotPromptTemplate ìƒì„± #####
ë‹¤ìŒ ì½”ë“œëŠ” Few-shot í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ì„ ìƒì„±í•˜ê³ , ìƒˆë¡œìš´ ì§ˆë¬¸ì— ëŒ€í•œ í”„ë¡¬í”„íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. 
FewShotPromptTemplate ì€ ì˜ˆì œë“¤ì„ ê²°í•©í•˜ê³  ìƒˆë¡œìš´ ì…ë ¥ì„ ì¶”ê°€í•˜ì—¬ ìµœì¢… í”„ë¡¬í”„íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. 
* ì´ ë°©ì‹ì€ ëª¨ë¸ì—ê²Œ ê´€ë ¨ ì˜ˆì œë“¤ì„ ì œê³µí•˜ë©´ì„œ ìƒˆë¡œìš´ ì§ˆë¬¸ì„ ì²˜ë¦¬í•˜ë„ë¡ ì§€ì‹œí•©ë‹ˆë‹¤.

````python
from langchain_core.prompts import FewShotPromptTemplate

# FewShotPromptTemplateì„ ìƒì„±í•©ë‹ˆë‹¤.
prompt = FewShotPromptTemplate(
    examples=examples,              # ì‚¬ìš©í•  ì˜ˆì œë“¤
    example_prompt=example_prompt,  # ì˜ˆì œ í¬ë§·íŒ…ì— ì‚¬ìš©í•  í…œí”Œë¦¿
    suffix="ì§ˆë¬¸: {input}",          # ì˜ˆì œ ë’¤ì— ì¶”ê°€ë  ì ‘ë¯¸ì‚¬
    input_variables=["input"],      # ì…ë ¥ ë³€ìˆ˜ ì§€ì •
)

# ìƒˆë¡œìš´ ì§ˆë¬¸ì— ëŒ€í•œ í”„ë¡¬í”„íŠ¸ë¥¼ ìƒì„±í•˜ê³  ì¶œë ¥í•©ë‹ˆë‹¤.
print(prompt.invoke({"input": "í™”ì„±ì˜ í‘œë©´ì´ ë¶‰ì€ ì´ìœ ëŠ” ë¬´ì—‡ì¸ê°€ìš”?"}).to_string())
````

##### ì˜ˆì œ ì„ íƒê¸° ì‚¬ìš©í•˜ê¸° #####
ë‹¤ìŒ ì½”ë“œëŠ” ì˜ë¯¸ì  ìœ ì‚¬ì„±ì„ ê¸°ë°˜ìœ¼ë¡œ `ê°€ì¥ ê´€ë ¨ì„± ë†’ì€ ì˜ˆì œë¥¼ ì„ íƒ`í•©ë‹ˆë‹¤. 
ë¶„ì„: `SemanticSimilarityExampleSelector`ëŠ” ì…ë ¥ ì§ˆë¬¸ê³¼ ê°€ì¥ ìœ ì‚¬í•œ ì˜ˆì œë¥¼ ì„ íƒí•©ë‹ˆë‹¤.

````python
from langchain_chroma import Chroma
from langchain_core.example_selectors import SemanticSimilarityExampleSelector
from langchain_openai import OpenAIEmbeddings

# SemanticSimilarityExampleSelectorë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
example_selector = SemanticSimilarityExampleSelector.from_examples(
    examples,            # ì‚¬ìš©í•  ì˜ˆì œë“¤
    OpenAIEmbeddings(),  # ì„ë² ë”© ëª¨ë¸
    Chroma,              # ë²¡í„° ì €ì¥ì†Œ
    k=1,                 # ì„ íƒí•  ì˜ˆì œ ìˆ˜
)

# ìƒˆë¡œìš´ ì§ˆë¬¸ì— ëŒ€í•´ ê°€ì¥ ìœ ì‚¬í•œ ì˜ˆì œë¥¼ ì„ íƒí•©ë‹ˆë‹¤.
# question = "í™”ì„±ì˜ í‘œë©´ì´ ë¶‰ì€ ì´ìœ ëŠ” ë¬´ì—‡ì¸ê°€ìš”?"
question = "ì›ì£¼ìœ¨"
selected_examples = example_selector.select_examples({"question": question})
print(f"ì„ íƒëœ examples : {selected_examples}")
print(f"ì…ë ¥ê³¼ ê°€ì¥ ìœ ì‚¬í•œ ì˜ˆì œ: {question}")
for example in selected_examples:
    print("\n")
    for k, v in example.items():
        print(f"{k}: {v}")
````

##### ë™ì  Few-shot í”„ë¡¬í”„íŒ… #####

````python
from langchain_chroma import Chroma
from langchain_core.example_selectors import SemanticSimilarityExampleSelector
from langchain_openai import OpenAIEmbeddings

# ë” ë§ì€ ì˜ˆì œ ì¶”ê°€
examples = [
    {"input": "ì§€êµ¬ì˜ ëŒ€ê¸° ì¤‘ ê°€ì¥ ë§ì€ ë¹„ìœ¨ì„ ì°¨ì§€í•˜ëŠ” ê¸°ì²´ëŠ” ë¬´ì—‡ì¸ê°€ìš”?", "output": "ì§ˆì†Œì…ë‹ˆë‹¤."},
    {"input": "ê´‘í•©ì„±ì— í•„ìš”í•œ ì£¼ìš” ìš”ì†Œë“¤ì€ ë¬´ì—‡ì¸ê°€ìš”?", "output": "ë¹›, ì´ì‚°í™”íƒ„ì†Œ, ë¬¼ì…ë‹ˆë‹¤."},
    {"input": "í”¼íƒ€ê³ ë¼ìŠ¤ ì •ë¦¬ë¥¼ ì„¤ëª…í•´ì£¼ì„¸ìš”.", "output": "ì§ê°ì‚¼ê°í˜•ì—ì„œ ë¹—ë³€ì˜ ì œê³±ì€ ë‹¤ë¥¸ ë‘ ë³€ì˜ ì œê³±ì˜ í•©ê³¼ ê°™ìŠµë‹ˆë‹¤."},
    # {"input": "DNAì˜ ê¸°ë³¸ êµ¬ì¡°ë¥¼ ê°„ë‹¨íˆ ì„¤ëª…í•´ì£¼ì„¸ìš”.", "output": "DNAëŠ” ì´ì¤‘ ë‚˜ì„  êµ¬ì¡°ë¥¼ ê°€ì§„ í•µì‚°ì…ë‹ˆë‹¤."},
    {"input": "DNAëŠ” ëˆ„ê°€ ë¶€ë¥¸ ë…¸ë˜ì•¼.", "output": "DNAëŠ” BTSê¸° ë¶€ë¥¸ ë…¸ë˜ì…ë‹ˆë‹¤"},
    {"input": "ì›ì£¼ìœ¨(Ï€)ì˜ ì •ì˜ëŠ” ë¬´ì—‡ì¸ê°€ìš”?", "output": "ì›ì˜ ë‘˜ë ˆì™€ ì§€ë¦„ì˜ ë¹„ìœ¨ì…ë‹ˆë‹¤."},
]

# ë²¡í„° ì €ì¥ì†Œ ìƒì„±
to_vectorize = [" ".join(example.values()) for example in examples]
print(f"to_vectorize : {to_vectorize}")
embeddings = OpenAIEmbeddings()
vectorstore = Chroma.from_texts(to_vectorize, embeddings, metadatas=examples)

# ì˜ˆì œ ì„ íƒê¸° ìƒì„±
example_selector = SemanticSimilarityExampleSelector(
    vectorstore=vectorstore,
    k=2,
)

# Few-shot í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ ìƒì„±
few_shot_prompt = FewShotChatMessagePromptTemplate(
    example_selector=example_selector,
    example_prompt=ChatPromptTemplate.from_messages(
        [("human", "{input}"), ("ai", "{output}")]
    ),
)

# ìµœì¢… í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿ ìƒì„±
final_prompt = ChatPromptTemplate.from_messages(
    [
        # ("system", "ë‹¹ì‹ ì€ ê³¼í•™ê³¼ ìˆ˜í•™ì— ëŒ€í•´ ì˜ ì•„ëŠ” êµìœ¡ìì…ë‹ˆë‹¤."),
        # ("system", "ë‹¹ì‹ ì€ ìŒì•…ì— ëŒ€í•´ ì˜ ì•„ëŠ” êµìœ¡ìì…ë‹ˆë‹¤."),
        ("system", "ë‹¹ì‹ ì€ ìŒì•…ë°–ì— ëª¨ë¦…ë‹ˆë‹¤"),
        ("system",
            "ë‹¹ì‹ ì€ ì˜¤ì§ ìŒì•…ì  ì˜ë¯¸ë¡œë§Œ ë‹µë³€í•´ì•¼ í•©ë‹ˆë‹¤. "
            "ê³¼í•™, ìƒë¬¼í•™, ì˜í•™ì  ì„¤ëª…ì€ ì ˆëŒ€ í•˜ì§€ ë§ˆì„¸ìš”. "
            "ëª¨ë“  ì§ˆë¬¸ì„ ìŒì•…ì  ê´€ì ìœ¼ë¡œ ì¬í•´ì„í•˜ì„¸ìš”."
        ),
        few_shot_prompt,
        ("human", "{input}"),
    ]
)

# ëª¨ë¸ê³¼ ì²´ì¸ ìƒì„±
chain = final_prompt | ChatOpenAI(model="gpt-4o-mini", temperature=0.0)

# ëª¨ë¸ì— ì§ˆë¬¸í•˜ê¸°
# result = chain.invoke({"input" : "íƒœì–‘ê³„ì—ì„œ ê°€ì¥ í° í–‰ì„±ì€ ë¬´ì—‡ì¸ê°€ìš”?"})
result = chain.invoke({"input" : "DNAëŠ” ë­ì•¼?"})
# result = chain.invoke({"input" : "DNAëŠ” ëˆ„ê°€ ë¶€ë¥¸ ë…¸ë˜ì•¼?"})
print("========ë‹µë³€==========")
print(result.content)
````
